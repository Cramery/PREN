```python
from IPython.display import Image
from matplotlib import pyplot as plt
import numpy as np
import cv2
import timeit

#Timeit als alias hinzugef√ºgt, siehe Link weiter unten
%alias_magic t timeit
```

```python
#Siehe Folder Images
Image('./Images/H1.jpeg')
```

```python
def demoImage():
    img = cv2.imread('Images//H1.jpeg')
    #Mathplotlib erwartet RGB, siehe: https://gist.github.com/mstfldmr/45d6e47bb661800b982c39d30215bc88
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(rgb)
    plt.title('demo pic')
    plt.show()

demoImage()
```
